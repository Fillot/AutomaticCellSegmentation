```python
root_folder = "/media/tom/SMART-1TO/Dani"
```

```python
red = 2 # meaning:  the red channel is the second one
green = 3
blue = 1 # note: blue is assumed to be DAPI

# the following values generally work well 
# for 1024x1024 at x60 resolution. 
# unit is pixels. 
# If yours significantly different, measure them in fiji
cell_diameter = 250 
nucleus_dimater = 150

```
